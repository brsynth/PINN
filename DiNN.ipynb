## ODE system

A general model for bacterial physiology is as follows:

\begin{equation}
\begin{split}
\frac{d[R]}{dt} &=
\Bigg(\frac{1}{\frac{1}{k_{\text{tr}}^{\text{init}}}+\frac{1}{k_{\text{tr}}^{\text{elong}}}}\Bigg)
\left(
\frac{1}{1+\frac{K_{\text{pol}}}{[\text{pol}]}
\prod_{i^{\text{rep}}}^{n^{\text{rep}}} \bigg( 1+\frac{[T^{\text{rep}}]}{K_T^{\text{rep}}} \bigg) \prod_{i^{\text{act}}}^{n^{\text{act}}} \Bigg( \frac{1+\frac{[T^{\text{act}}]}{K_T^{\text{act}}}e^{-\beta \epsilon}}{1+\frac{[T^{\text{act}}]}{K_T^{\text{act}}}} \Bigg)
^{-1}}
\rule{0cm}{1cm}\right)
-{k^{\text{rna}}_{\text{deg}}}[R]\\
\frac{d[P]}{dt} &= \Bigg(\frac{1}{\frac{1}{k_{\text{tl}}^{\text{init}}}+\frac{1}{k_{\text{tl}}^{\text{elong}}}}\Bigg)\Bigg(\frac{1}{1+\frac{K_{\text{rib}}}{[\text{rib}]}}\Bigg)[R]-{k^{\text{prot}}_{\text{deg}}}[P]\\
\frac{d[M]}{dt} &= (k_{\text{tp}}^{\text{passive}}+k_{\text{tp}}^{\text{active}})[C_{\text{eff}}]+\sum_{i^{+}}^{n^{+}}\frac{k_{\text{cat}}^{i^+}[E^+][M^+]}{k_{\text{M}}^{i^+} + [M^+] + [E^+]}-\sum_{i^{-}}^{n^{-}}\frac{k_{\text{cat}}^{i^-}[E^-][M^-]}{k_{\text{M}}^{i^-} + [M^-] + [E^-]}-(k_{\text{deg}}^{\text{met}}+k_{\text{exc}}^{\text{met}})[M]
\end{split}
\end{equation}


Here, we instantiate a minimal version of this model, with the 10 following species:
- 2 metabolites ($M_a, M_b$),<Br>
- 4 mRNAs and 4 associated proteins:
    - a metabolic enzyme ($R_e, P_e$),
    - a repressing transcription factor ($R_t, P_t$),
    - a ribosome ($R_r, P_r$),
    - a RNAP ($R_p, P_p$).

This minimal, simplified and non-dimensionalized model takes the following form:

\begin{equation}
\begin{split}

\frac{d[R_e]}{d\tau} &=
k_{\text{tr}}^e
\left(
\frac{1}{1+\frac{K_{\text{pol}}^e}{[P_p]}
\bigg( 1+\frac{[P_t]}{K_T^{\text{rep}}} \bigg)}
\rule{0cm}{1cm}\right)
-[R_e]\\

\frac{d[R_t]}{d\tau} &=
k_{\text{tr}}^t
\Bigg(
\frac{1}{1+\frac{K_{\text{pol}}^t}{[P_p]}}
\Bigg)
-{k^{t}_{\text{rdeg}}}[R_t]\\

\frac{d[R_r]}{d\tau} &=
k_{\text{tr}}^r
\Bigg(
\frac{1}{1+\frac{K_{\text{pol}}^r}{[P_p]}}
\Bigg)
-{k^{r}_{\text{rdeg}}}[R_r]\\

\frac{d[R_p]}{d\tau} &=
k_{\text{tr}}^p
\Bigg(
\frac{1}{1+\frac{K_{\text{pol}}^p}{[P_p]}}
\Bigg)
-{k^{p}_{\text{rdeg}}}[R_p]\\

\\

\frac{d[P_e]}{d\tau} &=
k_{\text{tl}}^{e}
\Bigg(\frac{1}{1+\frac{K^{e}_{\text{rib}}}{[P_r]}}\Bigg)[R_e]
-{k^{e}_{\text{pdeg}}}[P_e]\\

\frac{d[P_t]}{d\tau} &=
k_{\text{tl}}^{t}
\Bigg(\frac{1}{1+\frac{K^{t}_{\text{rib}}}{[P_r]}}\Bigg)[R_t]
-{k^{t}_{\text{pdeg}}}[P_t]\\

\frac{d[P_r]}{d\tau} &=
k_{\text{tl}}^{r}
\Bigg(\frac{1}{1+\frac{K^{r}_{\text{rib}}}{[P_r]}}\Bigg)[R_r]
-{k^{r}_{\text{pdeg}}}[P_r]\\

\frac{d[P_p]}{d\tau} &=
k_{\text{tl}}^{p}
\Bigg(\frac{1}{1+\frac{K^{p}_{\text{rib}}}{[P_r]}}\Bigg)[R_p]
-{k^{p}_{\text{pdeg}}}[P_p]\\

\\

\frac{d[M_a]}{d\tau} &=
k_{\text{tp}}
-\frac{k_{\text{cat}}[P_e][M_a]}{k_{\text{M}} + [M_a] + [P_e]}
-k_{\text{exc}}^{a}[M_a]\\

\frac{d[M_b]}{d\tau} &= \frac{k_{\text{cat}}[P_e][M_a]}{k_{\text{M}} + [M_a] + [P_e]}
-k_{\text{exc}}^{b}[M_b]

\end{split}
\end{equation}

The non-dimensionalization was done by: $\hat{k}^{e}_{\text{rdeg}}$, such that:<Br>
${k^{x}_{\text{rdeg}}}=\frac{{\hat{k}^{x}_{\text{rdeg}}}}{\hat{k}^{e}_{\text{rdeg}}}, 
{k^{x}_{\text{pdeg}}}=\frac{{\hat{k}^{x}_{\text{pdeg}}}}{\hat{k}^{e}_{\text{rdeg}}},
{k^{x}_{\text{tr}}}=\frac{{\hat{k}^{x}_{\text{tr}}}}{\hat{k}^{e}_{\text{rdeg}}},
{k^{x}_{\text{tl}}}=\frac{{\hat{k}^{x}_{\text{tl}}}}{\hat{k}^{e}_{\text{rdeg}}},
k_{\text{cat}}=\frac{{\hat{k}_{\text{cat}}}}{\hat{k}^{e}_{\text{rdeg}}},
k_{\text{tp}}=\frac{{\hat{k}_{\text{tp}}}}{\hat{k}^{e}_{\text{rdeg}}},
k_{\text{exc}}^{x}=\frac{{\hat{k}_{\text{exc}}^{x}}}{\hat{k}^{e}_{\text{rdeg}}},
d\tau=dt\cdot{\hat{k}^{e}_{\text{rdeg}}}
$

The list of 30 parameters is:<Br>
$k_{\text{tr}}^e, k_{\text{tr}}^t, k_{\text{tr}}^p, k_{\text{tr}}^r,$<Br>
$K_{\text{pol}}^e, K_{\text{pol}}^t, K_{\text{pol}}^p, K_{\text{pol}}^r,$<Br>
${k^{t}_{\text{rdeg}}}, {k^{t}_{\text{rdeg}}}, {k^{t}_{\text{rdeg}}}, {k^{t}_{\text{rdeg}}},$<Br>
$K_T^{\text{rep}},$<Br>
$k_{\text{tl}}^{e}, k_{\text{tl}}^{t}, k_{\text{tl}}^{p}, k_{\text{tl}}^{r},$<Br>
$K^{e}_{\text{rib}}, K^{t}_{\text{rib}}, K^{p}_{\text{rib}}, K^{r}_{\text{rib}},$<Br>
$k^{e}_{\text{pdeg}}, k^{t}_{\text{pdeg}}, k^{p}_{\text{pdeg}}, k^{r}_{\text{pdeg}},$<Br>
$k_{\text{tp}}, k_{\text{cat}}, k_{\text{M}}, k_{\text{exc}}^{a}, k_{\text{exc}}^{b}$

The model is further simplified as follows:
$k_{\text{tr}} = k_{\text{tr}}^e= k_{\text{tr}}^t= k_{\text{tr}}^p= k_{\text{tr}}^r$<Br>
$K_{\text{pol}} = K_{\text{pol}}^e= K_{\text{pol}}^t= K_{\text{pol}}^p= K_{\text{pol}}^r$<Br>
${k_{\text{rdeg}}} = {k^{t}_{\text{rdeg}}} = {k^{t}_{\text{rdeg}}} = {k^{t}_{\text{rdeg}}} = {k^{t}_{\text{rdeg}}}$<Br>
$k_{\text{tl}} = k_{\text{tl}}^{t} = k_{\text{tl}}^{p}= k_{\text{tl}}^{r}$<Br>
$K_{\text{rib}} = K^{e}_{\text{rib}} = K^{t}_{\text{rib}} = K^{p}_{\text{rib}}= K^{r}_{\text{rib}}$<Br>
$k_{\text{pdeg}} = k^{e}_{\text{pdeg}} = k^{t}_{\text{pdeg}}= k^{p}_{\text{pdeg}}= k^{r}_{\text{pdeg}} \approx 0$<Br>
$k_{\text{exc}} = k_{\text{exc}}^{a} = k_{\text{exc}}^{b}$

As such, there are 10 remaining parameters in this simplified model:<Br>
$k_{\text{tr}}, K_{\text{pol}}, {k_{\text{rdeg}}}, k_{\text{tl}}, K_{\text{rib}}, k_{\text{exc}},$<Br>
$K_T^{\text{rep}},k_{\text{tp}}, k_{\text{cat}}, k_{\text{M}}$

With the following set of equations:

\begin{equation}
\begin{split}

\frac{d[R_e]}{d\tau} &=
k_{\text{tr}}
\left(
\frac{1}{1+\frac{K_{\text{pol}}}{[P_p]}
\bigg( 1+\frac{[P_t]}{K_T^{\text{rep}}} \bigg)}
\rule{0cm}{1cm}\right)
-[R_e]\\

\frac{d[R_t]}{d\tau} &=
k_{\text{tr}}
\Bigg(
\frac{1}{1+\frac{K_{\text{pol}}}{[P_p]}}
\Bigg)
-{k_{\text{rdeg}}}[R_t]\\

\frac{d[R_r]}{d\tau} &=
k_{\text{tr}}
\Bigg(
\frac{1}{1+\frac{K_{\text{pol}}}{[P_p]}}
\Bigg)
-{k_{\text{rdeg}}}[R_r]\\

\frac{d[R_p]}{d\tau} &=
k_{\text{tr}}
\Bigg(
\frac{1}{1+\frac{K_{\text{pol}}}{[P_p]}}
\Bigg)
-{k_{\text{rdeg}}}[R_p]\\

\\

\frac{d[P_e]}{d\tau} &=
k_{\text{tl}}
\Bigg(\frac{1}{1+\frac{K_{\text{rib}}}{[P_r]}}\Bigg)[R_e]
\\

\frac{d[P_t]}{d\tau} &=
k_{\text{tl}}
\Bigg(\frac{1}{1+\frac{K_{\text{rib}}}{[P_r]}}\Bigg)[R_t]
\\

\frac{d[P_r]}{d\tau} &=
k_{\text{tl}}
\Bigg(\frac{1}{1+\frac{K_{\text{rib}}}{[P_r]}}\Bigg)[R_r]
\\

\frac{d[P_p]}{d\tau} &=
k_{\text{tl}}
\Bigg(\frac{1}{1+\frac{K_{\text{rib}}}{[P_r]}}\Bigg)[R_p]
\\

\\

\frac{d[M_a]}{d\tau} &=
k_{\text{tp}}
-\frac{k_{\text{cat}}[P_e][M_a]}{k_{\text{M}} + [M_a] + [P_e]}
-k_{\text{exc}}[M_a]\\

\frac{d[M_b]}{d\tau} &= \frac{k_{\text{cat}}[P_e][M_a]}{k_{\text{M}} + [M_a] + [P_e]}
-k_{\text{exc}}[M_b]


\end{split}
\end{equation}

```python
# Solving the ODE system
from scipy.integrate import odeint
from deriv_equations import deriv
import numpy as np

# A grid of time points (in days)
t = np.linspace(0, 10, 100) #from day 0 to day 500, generate 100 points

# Initial conditions
R_e_0 = 3
R_t_0 = 2
R_r_0 = 4
R_p_0 = 1

P_e_0 = 1
P_t_0 = 1
P_r_0 = 1
P_p_0 = 1

M_a_0 = 1
M_b_0 = 1

# Initial conditions vector
y0 = R_e_0, R_t_0 , R_r_0, R_p_0, P_e_0, P_t_0, P_r_0, P_p_0, M_a_0, M_b_0

# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(
    k_tr,
    K_pol,
    k_rdeg,
    k_tl,
    K_rib,
    k_exc,
    K_T_rep,
    k_tp,
    k_cat,
    k_M
))
R_e, R_t, R_r, R_p, P_e, P_t, P_r, P_p, M_a, M_b = ret.T
```

## Neural Network Training

```python
# Training parameters 
learning_rate = 1e-6
epoch_number = 100000
# epoch_number = 1000
Base_LR = 1e-5
Max_LR = 1e-3
Step_Size_Up = 1000
Mode = "exp_range"
Gamma = 0.85
Cycle_Momentum = False
```

```python
# Training
from dinn import DINN
true_stuff = [
k_tr,K_pol,k_rdeg,k_tl,K_rib,k_exc,K_T_rep,k_tp,k_cat,k_M
]

dinn = DINN(true_stuff,
            ranges,
            dinn_data[0],
            dinn_data[1],
            dinn_data[2],
            dinn_data[3], 
            dinn_data[4],
            dinn_data[5],
            dinn_data[6],
            dinn_data[7],
            dinn_data[8],
            dinn_data[9],
            dinn_data[10],
            ) #in the form of [t,S,I,D,R]

optimizer = optim.Adam(
    dinn.params,
    lr = learning_rate
    )
dinn.optimizer = optimizer

scheduler = torch.optim.lr_scheduler.CyclicLR(
    dinn.optimizer,
    base_lr=Base_LR,
    max_lr=Max_LR,
    step_size_up=Step_Size_Up,
    mode=Mode,
    gamma=Gamma,
    cycle_momentum=Cycle_Momentum
    )
dinn.scheduler = scheduler



r2_store, R_e_pred_list, R_t_pred_list, R_r_pred_list, R_p_pred_list, P_e_pred_list, P_t_pred_list, P_r_pred_list, P_p_pred_list, M_a_pred_list, M_b_pred_list = dinn.train(epoch_number) #train
```
