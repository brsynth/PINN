## Solving the ODE system for 1mM

```python
# Solving the ODE system for 1mM
import importlib
from scipy.integrate import odeint

import deriv_equations_Millard
importlib.reload(deriv_equations_Millard)
from deriv_equations_Millard import deriv_Millard

import Millard_dict
importlib.reload(Millard_dict)
from Millard_dict import DICT

import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt

t = np.linspace(0, 5, 10000) 

# Initial conditions
GLC_1_0 = 12.6597333333333
ACE_env_1_0 = 0.958577777777778
X_1_0 = 0.06308
ACCOA_1_0 = 0
ACP_1_0 = 0
ACE_cell_1_0 = 0

# Initial conditions vector
y_1_0 = [GLC_1_0, ACE_env_1_0, X_1_0, ACCOA_1_0, ACP_1_0, ACE_cell_1_0] 

ret = odeint(deriv_Millard, y_1_0, t, args=(DICT,))

GLC, ACE_env, X, ACCOA, ACP, ACE_cell = ret.T

# Load calibration data 
ode_data_1 = genfromtxt('data_1mM.csv', delimiter=',')

data_t = ode_data_1[1:, 0]  

# Plot everything
fig, axs = plt.subplots(1, 3, figsize=(15, 3))

axs[0].plot(t, GLC, label='GLC_cal', color='b')
axs[0].plot(data_t, ode_data_1[1:, 3], 'o', label='GLC_data', color='b')
axs[0].set_title('GLC 1mM')
axs[0].set_xlabel('t')
axs[0].set_ylabel('GLC')
axs[0].legend()
axs[0].grid(True)

axs[1].plot(t, ACE_env, label='ACE_env_cal', color='g')
axs[1].plot(data_t, ode_data_1[1:, 1], 'o', label='ACE_env_data', color='g')
axs[1].set_title('ACE 1mM')
axs[1].set_xlabel('t')
axs[1].set_ylabel('ACE_env')
axs[1].legend()
axs[1].grid(True)

axs[2].plot(t, X,label='X_cal', color='r')
axs[2].plot(data_t, ode_data_1[1:, 2], 'o',label='X_data', color='r')
axs[2].set_title('X 1mM')
axs[2].set_xlabel('t')
axs[2].set_ylabel('X')
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()

```
